```python
def fixClosedDate(date):
    if date is np.NaN:
        return datetime.datetime.today()
    else:
        return datetime.datetime.strptime(date,'%m/%d/%Y %I:%M:%S %p')
```

```python
#newdata = newdata[newdata['Closed Date'].notnull()]
```

```python
newdata.loc[:, 'Created Date'] = newdata.loc[:, 'Created Date'].apply(lambda x:datetime.datetime.strptime(x,'%m/%d/%Y %I:%M:%S %p'))
```

```python
newdata['Created Date'].head()
```

```python
newdata.loc[:, 'Closed Date'] = newdata.loc[:, 'Closed Date'].apply(fixClosedDate)
```

```python
newdata['Closed Date'].head()
```

```python
newdata.loc[:, 'processing_time'] =  newdata.loc[:, 'Closed Date'] - newdata.loc[:, 'Created Date']
```

```python
newdata.head()
```

```python
newdata.loc[:, 'start_time_window'] = newdata.loc[:, 'Created Date'].apply(lambda x: x.hour)
```

```python
newdata['start_time_window'].head()
```

```python
newdata.to_csv('output1.csv')
```

```python
newdata.loc[:, 'float_time'] =newdata.loc[:, 'processing_time'].apply(lambda x:x/np.timedelta64(1, 'D'))
```

```python
agency_borough = newdata[['Agency', 'Borough', 'float_time']].groupby(['Agency', 'Borough']).mean().unstack()
```

```python
agency_borough
```

```python
#We'll arrange the subplots in two rows and three columns. 
#Since we have only 5 boroughs, one plot will be blank
COL_NUM = 2
ROW_NUM = 3
import matplotlib.pyplot as plt
fig, axes = plt.subplots(ROW_NUM, COL_NUM, figsize=(12,12))

for i, (label, col) in enumerate(agency_borough.iteritems()): 
    ax = axes[int(i/COL_NUM), i%COL_NUM]
    col = col.sort_values(ascending=False)
    col.plot(kind='barh', ax=ax)
    ax.set_title(label)

plt.tight_layout() 
```
