```python
def fixClosedDate(date):
    if date is np.NaN:
        return datetime.datetime.today()
    else:
        return datetime.datetime.strptime(date,'%m/%d/%Y %I:%M:%S %p')
```

```python
#newdata = newdata[newdata['Closed Date'].notnull()]
```

```python
newdata.loc[:, 'Created Date'] = newdata.loc[:, 'Created Date'].apply(lambda x:datetime.datetime.strptime(x,'%m/%d/%Y %I:%M:%S %p'))
```

```python
newdata['Created Date'].head()
```

```python
newdata.loc[:, 'Closed Date'] = newdata.loc[:, 'Closed Date'].apply(fixClosedDate)
```

```python
newdata['Closed Date'].head()
```

```python
newdata.loc[:, 'processing_time'] =  newdata.loc[:, 'Closed Date'] - newdata.loc[:, 'Created Date']
```

```python
newdata.head()
```

```python
newdata.loc[:, 'start_time_window'] = newdata.loc[:, 'Created Date'].apply(lambda x: x.hour)
```
